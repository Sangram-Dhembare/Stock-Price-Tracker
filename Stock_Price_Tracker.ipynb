```python
import yfinance as yf
import matplotlib.pyplot as plt

def track_stock():
    # User inputs the ticker
    ticker = input("Enter Stock Ticker (e.g., AAPL, TSLA, BTC-USD): ").upper()

    # Download data (Last 6 months)
    data = yf.download(ticker, period="6mo", interval="1d")

    if data.empty:
        print("Error: Could not find that ticker.")
        return

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], color='#1f77b4', linewidth=2)

    plt.title(f"{ticker} - 6 Month Price Trend", fontsize=16)
    plt.ylabel("Price (USD)")
    plt.grid(True, alpha=0.3)

    # This line ensures the plot shows up inside Colab
    plt.show()

track_stock()
```
